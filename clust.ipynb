```python
q1_pengeluaran = df_mean['Pengeluaran Per Kapita Riil Disesuaikan (Ribu Rupiah)'].quantile(0.25)
q3_pengeluaran = df_mean['Pengeluaran Per Kapita Riil Disesuaikan (Ribu Rupiah)'].quantile(0.75)

def categorize_pengeluaran(value):
    if value <= q1_pengeluaran:
        return 'Kelas Ekonomi Rendah'
    elif value <= q3_pengeluaran:
        return 'Kelas Ekonomi Menengah'
    else:
        return 'Kelas Ekonomi Tinggi'

df_mean['Kategori_Pengeluaran'] = df_mean['Pengeluaran Per Kapita Riil Disesuaikan (Ribu Rupiah)'].apply(categorize_pengeluaran)

q1_penduduk = df_mean['Jumlah Penduduk (Ribu)'].quantile(0.25)
q3_penduduk = df_mean['Jumlah Penduduk (Ribu)'].quantile(0.75)

def categorize_penduduk(value):
    if value <= q1_penduduk:
        return 'Populasi Sedikit'
    elif value <= q3_penduduk:
        return 'Populasi Sedang'
    else:
        return 'Populasi Banyak'

df_mean['Kategori_Penduduk'] = df_mean['Jumlah Penduduk (Ribu)'].apply(categorize_penduduk)
q1_luas = df_mean['Luas Wilayah'].quantile(0.25)
q3_luas = df_mean['Luas Wilayah'].quantile(0.75)

def categorize_luas(value):
    if value <= q1_luas:
        return 'Wilayah Kecil'
    elif value <= q3_luas:
        return 'Wilayah Sedang'
    else:
        return 'Wilayah Besar'

df_mean['Kategori_Luas_Wilayah'] = df_mean['Luas Wilayah'].apply(categorize_luas)

display(df_mean[['Kabupaten/Kota', 'Kategori_Pengeluaran', 'Kategori_Penduduk', 'Kategori_Luas_Wilayah']].head())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)

df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Kabupaten/Kota"] = df_mean["Kabupaten/Kota"]

df_pca.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, cluster_labels)
    silhouette_scores.append(score)

best_k = K_range[silhouette_scores.index(max(silhouette_scores))]
best_score = max(silhouette_scores)

print(f"K terbaik: {best_k}")
print(f"Silhouette Score terbaik: {best_score:.4f}")

plt.figure(figsize=(8, 5))
plt.plot(list(K_range), silhouette_scores, marker='o', linestyle='-')
plt.title("Silhouette Score untuk K=2 hingga K=10 (Setelah PCA)")
plt.xlabel("Jumlah Cluster (K)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca_final = pca.fit_transform(X_normalized)

final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
clusters = final_kmeans.fit_predict(X_pca_final)

df_mean["Cluster"] = clusters
display(df_mean[["Kabupaten/Kota", "Cluster"]].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame for visualization with the first two PCA components and cluster labels
df_viz = pd.DataFrame(X_pca_final[:, :2], columns=["PC1", "PC2"])
df_viz["Cluster"] = clusters
df_viz["Kabupaten/Kota"] = df_mean["Kabupaten/Kota"]

plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df_viz, x="PC1", y="PC2",
    hue="Cluster", palette="viridis", s=100
)

# Add text labels for each point
for i in range(df_viz.shape[0]):
    plt.text(df_viz["PC1"][i] + 0.1, df_viz["PC2"][i] + 0.1, df_viz["Kabupaten/Kota"][i], fontsize=8)

plt.title("Visualisasi Clustering K-Means (PCA)")
plt.xlabel("Principal Component 1 (PC1)")
plt.ylabel("Principal Component 2 (PC2)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
